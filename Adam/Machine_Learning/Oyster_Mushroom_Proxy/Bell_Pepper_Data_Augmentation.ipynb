```python
def create_vertical_data():
  var = "Vertical_Flip_"
  for i in classes:
    path = os.path.join(directory, i)   # directory path for bell pepper images
    class_num = classes.index(i)
    for img in os.listdir(path):
      img_array = cv2.flip(cv2.imread(os.path.join(path,img)), 0)
      # img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
      # img_array = cv2.resize(img_array, (416, 416), interpolation=cv2.INTER_CUBIC)                    # Enlarging original image. INTER_CUBIC is best for this
      filename = var + img
      cv2.imwrite(filename, img_array)
      # print(img_array.shape)
      print(filename)
      


create_vertical_data()
```

Gaussian Noise


```python
cd '/content/drive/Shareddrives/ECEN 404 Machine Learning/Color/Final Model Bell Pepper/Dataset/Gaussian_Noise_Augmentation/Bacterial Spot'
```

```python
def create_gaussian_noise_data():
  var = "Gaussian_Noise_"
  for i in classes:
    path = os.path.join(directory, i)   # directory path for rice images
    class_num = classes.index(i)
    for img in os.listdir(path):
      img_array = cv2.imread(os.path.join(path,img))
      img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
      img_array = skimage.util.random_noise(img_array, mode='gaussian')
      # img_array = cv2.resize(img_array, (112, 112), interpolation=cv2.INTER_AREA)                    # Multiple of 32, Shrinking original image to 224x224 which is the reason I used INTER_AREA for the interpolation
      filename = var + img

      # plt.imshow(img_array)
      # plt.show()
      # cv2.imwrite(filename, img_array)
      skimage.io.imsave(filename, img_array)
      print(filename)

create_gaussian_noise_data()
```

Shearing

```python
cd '/content/drive/Shareddrives/ECEN 404 Machine Learning/Color/Final Model Bell Pepper/Dataset/Shearing_Augmentation/Bacterial Spot'
```

```python
ls
```

```python
def create_shearing_data():
  X_shear=[]
  y_shear=[]
  width = 256
  height = 256
  M2 = np.float32([[1, 0, 0], [0.2, 1, 0]])
  M2[0,2] = -M2[0,1] * width/2
  M2[1,2] = -M2[1,0] * height/2
  var = "Shearing_"
  for i in classes:
    path = os.path.join(directory, i)   # directory path for rice images
    class_num = classes.index(i)
    for img in os.listdir(path):
      img_array = cv2.imread(os.path.join(path,img))
      # img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
      img_array = cv2.warpAffine(img_array, M2, (width, height))
      # img_array = cv2.resize(img_array, (112, 112), interpolation=cv2.INTER_AREA)                    # Multiple of 32, Shrinking original image to 224x224 which is the reason I used INTER_AREA for the interpolation
      filename = var + img
      cv2.imwrite(filename, img_array)
      print(filename)


create_shearing_data()
```

```python
def create_rotate_before_flip_data():
  rotations = [30, 45, 60, 300, 315, 330]
  var = "Rotate_Before_Flip_"
  for i in classes:
    path = os.path.join(directory, i)   # directory path for bell pepper images
    class_num = classes.index(i)
    for img in os.listdir(path):
      j = random.randint(0,5)
      img_array = cv2.imread(os.path.join(path,img))
      img_array = imutils.rotate(img_array, angle=rotations[j])
      # img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
      # img_array = cv2.resize(img_array, (416, 416), interpolation=cv2.INTER_CUBIC)                    # Enlarging original image. INTER_CUBIC is best for this

      filename = var + img
      cv2.imwrite(filename, img_array)
      # print(filename)

create_rotate_before_flip_data()
```

Gaussian Blur

```python
cd '/content/drive/Shareddrives/ECEN 404 Machine Learning/Color/Final Model Bell Pepper/Dataset/Gaussian_Blur/Bacterial Spot'
```

```python
def create_gaussian_blur_data():
  blur = [9, 15]
  var = "Gaussian_Blur_"
  for i in classes:
    path = os.path.join(directory, i)   # directory path for bell pepper images
    class_num = classes.index(i)
    for img in os.listdir(path):
      j = random.randint(0,1)
      blur_val = blur[j]
      img_array = cv2.imread(os.path.join(path,img))
      img_array = cv2.GaussianBlur(img_array, (blur_val, blur_val), cv2.BORDER_DEFAULT)
      # img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
      # img_array = cv2.resize(img_array, (416, 416), interpolation=cv2.INTER_CUBIC)                    # Enlarging original image. INTER_CUBIC is best for this

      filename = var + img
      cv2.imwrite(filename, img_array)
      # print(filename)

create_gaussian_blur_data()
```
