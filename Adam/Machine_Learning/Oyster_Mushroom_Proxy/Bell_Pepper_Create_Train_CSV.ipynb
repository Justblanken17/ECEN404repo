```python
train_csv = []
filename = []
label = []

for i in os.listdir(train_directory):
  for j in classes:
    path = os.path.join(train_directory, i, j)
    class_num = classes.index(j)
    for img in os.listdir(path):
      print(img)
      filename.append(img)
      label.append(class_num)
train_csv.append(filename)
train_csv.append(label)
train_csv = np.array(train_csv)
print(type(train_csv))
train_csv = train_csv.T
train_csv = pd.DataFrame(train_csv)
train_csv = train_csv.sample(frac=1)
train_csv.columns = ['filename', 'labels']
print(train_csv.head())
train_csv.to_csv('train_labels.csv', index=False)
```
